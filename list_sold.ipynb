```python
# Initial imports
import os
import requests
import pandas as pd
from dotenv import load_dotenv
import datetime

%matplotlib inline
```

```python
# Load .env environment variables
load_dotenv()
```

```python
# Set Realtor API key 
api_key = os.getenv("RAPID_API_KEY")
```

```python
url = "https://realtor.p.rapidapi.com/properties/v2/list-sold"

querystring = {"state_code":"CA","prop_type":"single_family,multi_family,condo","sort":"sold_date","offset":"0"}

headers = {
    'x-rapidapi-key': api_key,
    'x-rapidapi-host': "realtor.p.rapidapi.com"
    }

response = requests.request("GET", url, headers=headers, params=querystring)
```

```python
limit = 200
ret = []
for x in range(50):
    querystring = {"state_code":"CA","prop_type":"single_family,multi_family,condo","sort":"sold_date","offset":x*limit}
    response = requests.request("GET", url, headers=headers, params=querystring)
    real_estate_data = response.json()
    re_df = real_estate_data['properties']
    ret.append(re_df)
```
