```python
# Initial imports
import os
import requests
import pandas as pd
from dotenv import load_dotenv
import datetime

%matplotlib inline
```

```python
# Load .env environment variables
load_dotenv()
```

```python
# Set Realtor API key 
api_key = os.getenv("RAPID_API_KEY")
```

```python
url = "https://realtor.p.rapidapi.com/properties/v2/list-for-rent"

querystring = {"state_code":"CA","prop_type":"single_family,multi_family,condo","sort":"sold_date","offset":"0"}

headers = {
    'x-rapidapi-key': api_key,
    'x-rapidapi-host': "realtor.p.rapidapi.com"
    }

response = requests.request("GET", url, headers=headers, params=querystring)
```

```python
# Convert JSON string to a dictionary
real_estate_data = response.json()
```

```python
# Create Dataframe 
re_df = real_estate_data['properties']

real_estate_df = pd.json_normalize(re_df)[['prop_type', 'list_date','last_update', 'prop_status', 'year_built', 'address.city', 'address.county','address.postal_code','address.lat','address.lon','community.price_min', 'community.baths_min', 'community.beds_min', 'community.sqft_min']]
real_estate_df.to_csv('list_for_rent_data.csv')
real_estate_df
```
